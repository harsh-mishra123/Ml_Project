# 🏥 HealthRisk_AI — Exploratory Data Analysis

This notebook provides an interactive exploration of the health risk dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid', palette='viridis')
%matplotlib inline
```

```python
df = pd.read_csv('../data/health_data.csv')
print(f'Shape: {df.shape}')
df.head(10)
```

## 1. Basic Info

```python
df.info()
print('\nMissing values:')
print(df.isnull().sum())
```

```python
df.describe()
```

## 2. Target Distribution

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['risk_level'].value_counts().plot(kind='bar', color=['#2ecc71', '#f39c12', '#e74c3c'], ax=ax)
ax.set_title('Risk Level Distribution')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 3. Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 4. Feature Distributions

```python
numeric_cols = numeric_df.columns.tolist()
fig, axes = plt.subplots(3, 4, figsize=(16, 10))
axes = axes.flatten()
for i, col in enumerate(numeric_cols):
    if i < len(axes):
        sns.histplot(df[col], kde=True, ax=axes[i], color='steelblue')
        axes[i].set_title(col)
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Feature Distributions', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

## 5. Box Plots by Risk Level

```python
key_features = ['age', 'bmi', 'blood_pressure_systolic', 'cholesterol', 'blood_glucose']
fig, axes = plt.subplots(1, len(key_features), figsize=(18, 5))
for i, col in enumerate(key_features):
    sns.boxplot(data=df, x='risk_level', y=col, ax=axes[i], palette='Set2', order=['Low', 'Medium', 'High'])
    axes[i].set_title(col)
plt.suptitle('Key Features by Risk Level', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

## 6. Pair Plot

```python
sns.pairplot(df[key_features + ['risk_level']], hue='risk_level', palette='Set1', diag_kind='kde')
plt.suptitle('Pair Plot — Key Features', y=1.02)
plt.show()
```
